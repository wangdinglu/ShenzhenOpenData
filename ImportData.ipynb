```python
def dataToDatabase(dataSource,classType,name):
    for r in dataSource.index:
        item = classType()
        item['_id'] = uuid.uuid4()
        for z in dataSource.columns:
            if type(dataSource.loc[r,z]) == np.int64:
                item[z] = int(dataSource.loc[r,z])
            else:
                item[z] = dataSource.loc[r,z]
        db[name].insert_one(item)

        if r % (len(dataSource.index) / 100) == 0:
            print('Progress: ', str(r / len(dataSource.index) * 100),'%')
```

```python
def InsertCompanyPosition(dataSource,start,name):
    for r in dataSource.index:
        if r > start:
            item = {}
            insert = False
            for z in dataSource.columns:
                if z == '企业公示的地址':
                    if type(dataSource.loc[r,z])==str:
                        feedback = requests.get('http://api.map.baidu.com/geocoding/v3/?address={}&output=json&ak=ey8QSVeHGHrlUGv3gWML7VKA0D6Q0wwT&callback=showLocation'.format(dataSource.loc[r,z]))
                        info = ast.literal_eval(feedback.text.replace('showLocation&&showLocation',''))
                        if info['status'] == 0:
                            lng = float(info['result']['location']['lng'])
                            lat = float(info['result']['location']['lat'])
                            transfer = requests.get('http://api.map.baidu.com/geoconv/v1/?coords={},{}&from=1&to=5&ak=ey8QSVeHGHrlUGv3gWML7VKA0D6Q0wwT'.format(lng,lat))
                            info = ast.literal_eval(transfer.text)
                            lng = float(info['result'][0]['x'])
                            lat = float(info['result'][0]['y']) 
                            item[companyIndex[z]] = dataSource.loc[r,z]
                            item['Coodrinates']=[lng,lat]
                            insert = True
                elif z == '成立日期':
                    item[companyIndex[z]] = datetime.strptime(dataSource.loc[r,z], '%Y-%m-%d')
                else:
                    item[companyIndex[z]] = dataSource.loc[r,z]
            item['Subtype']=name
            if insert:
                db['PingshanCompany'].insert_one(item)
            if r % int(len(dataSource.index)/50) == 0:
                print('Progress: ' + str(r) + ' Total: ' + str(len(dataSource.index)))
```

```python
def InsertHousePosition(dataSource):
    for r in dataSource.index:
        item = {}
        insert = False
        for z in dataSource.columns:
            if z == '小区':
                if type(dataSource.loc[r,z])==str:
                    feedback = requests.get('http://api.map.baidu.com/geocoding/v3/?address=深圳坪山{}&output=json&ak=ey8QSVeHGHrlUGv3gWML7VKA0D6Q0wwT&callback=showLocation'.format(dataSource.loc[r,z]))
                    info = ast.literal_eval(feedback.text.replace('showLocation&&showLocation',''))
                    if info['status'] == 0:
                        lng = float(info['result']['location']['lng'])
                        lat = float(info['result']['location']['lat'])
                        transfer = requests.get('http://api.map.baidu.com/geoconv/v1/?coords={},{}&from=1&to=5&ak=ey8QSVeHGHrlUGv3gWML7VKA0D6Q0wwT'.format(lng,lat))
                        info = ast.literal_eval(transfer.text)
                        lng = float(info['result'][0]['x'])
                        lat = float(info['result'][0]['y']) 
                        item[houseIndex[z]] = dataSource.loc[r,z]
                        item['geomerty']={'type':"Point",'coodrinates':[lng,lat]}
                        insert = True
            elif z == '面积':
                item[houseIndex[z]] = float(dataSource.loc[r,z].replace('平米',''))
            elif z == '总价':
                item[houseIndex[z]] = float(dataSource.loc[r,z])
            elif z == '单价':
                item[houseIndex[z]] = float(dataSource.loc[r,z])
            else:
                item[houseIndex[z]] = dataSource.loc[r,z]
        if insert:
            db['PingshanErshoufang'].insert_one(item)
        if r % int(len(dataSource.index)/50) == 0:
            print('Progress: ' + str(r) + ' Total: ' + str(len(dataSource.index)))
```
