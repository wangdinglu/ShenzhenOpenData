```python
import pymongo
client = pymongo.MongoClient('localhost', 27017)
db = client['ShenzhenOpenData']
```

```python
points = db.PingshanPOI.find()
index = 0
for p in points:
    index += 1
    db.PingshanPOI.update_one({'_id':p['_id']},{'$set': {'geometry': {'type':'Point','coordinates':p['coordinates']}}})
    if index%500==0:
        print(index)
```

```python
query = {"block_id":"e56e524a-261b-4bc7-ae4a-b439460783a6"}
block = db.PingshanCommunity

.find_one(query)
print(block["coordinates"])
points = db.PingshanPOI.find({
    "geometry.coordinates":{
        "$geoWithin":{
            "$geometry": {
             "type" : "Polygon" ,
             "coordinates": [block["coordinates"]]
            }
        }
    }
})

for p in points:
    print(p)
```

```python
blocks = db.PingshanStreet.find()
index = 0
for b in blocks:
    index += 1
    coordinates = []
    for pos in b['coordinates'][0]:
        coordinates.append([float(pos[0]),float(pos[1])])
        
    db.PingshanStreet.update_one({'_id':b['_id']},{'$set': {'coordinates':coordinates}})
    if index%100==0:
        print(index)
```
